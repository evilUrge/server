```python
from mining import init_remote_api, batch_iter
init_remote_api("hebkhan")

from google.appengine.ext import db
import models


```

```python
all_vids = list(batch_iter(models.Video.all()))
```

```python
version = models.TopicVersion.all().filter("edit = ", True).get()

```

```python
def add_change(content, **kw):
    change = models.VersionContentChange(parent = version)
    change.version = version
    change.content = content
    change.content_changes = kw
    return change
```

```python
import csv
r=csv.reader(open("/home/oferko/hebrew-khan/application/ani10.csv"))
from collections import namedtuple
V = namedtuple("V", next(r))
ani10 = {v.youtube.partition("?v=")[-1][:11]:v for v in (V(*row) for row in r)}
changes = []
for v in all_vids:
    av = ani10.get(v.youtube_id)
    if av and v.title.encode("utf8") != av.title:
        changes.append(add_change(v, title=av.title.decode("utf8")))
        print av.title
```

```python
for i in xrange(5, len(changes), 50):
    x=db.put(changes[i:i+50])
    print len(x)
```
